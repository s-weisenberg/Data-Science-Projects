__Step 1: Importing the Dataset into a Pandas Dataframe__

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('kag_risk_factors_cervical_cancer.csv')
```

__Step 2: Initial Dataset Exploration__

```python
df.head(10)
```

Here we can see the dataset has 36 columns, with missing values marked as '?'. The last 4 columns, 'Hinselmann', 'Schiller', 'Citology', 'Biopsy', hold the target variables. All attributes appear to be either numerical or boolean (0/1 binary).

```python
df.info()
```

From this summary, the number of rows (patients) is 858. Also, it is likely that all columns listed as "non-null object" type contain some missing values, indicated by the '?' character. This will be tested and verified in the next step.

__Step 3: Marking Missing Values as NaN__<br>
In order to correctly identify missing values in the data, the '?' character must be replaced with the standard NaN. This will allow for more informative exploration of the extent of the missing data. 

```python
df = df.replace('?', np.nan)
df.head(10)
```

```python
df.info()
```

There are still some columns with mixed or 'object' types. These must be transformed to numeric values (int or float) in order to use them for statistical or machine learning purposes.

```python
for column, values in df.iteritems():
        df[column] = pd.to_numeric(df[column], errors='coerce')
        
df.info()
```

__Step 4: Handling Missing Data__<br>
In order to work with a complete dataset, the missing values will be replaced by their medians for each column respectively. This may result in some skewing, particularly for the binary attributes (for which the median is equal to the mode). The merits of this technique will be tested later on in evaluation of performance metrics for the model against a model in which missing data is not replaced (ommitted or kept). 

```python
df_median_imputed = df.fillna(df.median())
df_median_imputed.head()
```
